### (1)주제 선정 이유: 
주제: 소주와 맥주는 대체재인가?<br>
소주와 맥주의 취향 차이가 갈리는 경우를 보고 그 둘 대체재인지 알고 싶었다.<br>
국제대학 전공 수업 Microeconomics 에서 A재화의 가격이 오르면 A재화의 소비는 줄고 대체재인 B재화의 소비는 늘어난다고 배웠다. <br>
가격 자료를 얻을 수 없어 세금 자료로 대체했다.<br>
세금이 오르면 가격이 오른 것으로 간주한다.

### (2)가설 정의: 
소주와 맥주는 대체재이다. <br>
따라서 소주 가격이 오르면<br>
즉, 소주에 부과되는 세금이 오르면 
소주 판매량은 줄고 맥주 판매량은 늘어날 것이다. <br>
반대로 맥주 가격이 오르면 <br>
즉, 맥주에 부과되는 세금이 오르면
맥주 판매량은 줄고 소주 판매량은 늘어날 것이다.

### (3)인터넷을 통한 데이터 획득: 
공공 데이터 포털에서 여러 주류에 부과된 세금과 주류 판매량 csv파일을 접했다.<br>
pandas에서 제공하는 read_csv로 두 파일을 읽었다. <br>

### (4)분석을 위한 데이터의 가공: 
위스키, 과실주 등 여러 주류가 포함된 자료에서 내가 필요한 '연도','맥주','소주'의 자료만을 취했다.<br>
'맥주', '소주'로만 적혀있는 카테고리를 구분하기 쉽게 '맥주 부과세금','맥주 판매량'등으로 다시 이름 붙였다.<br>
세금 자료와 판매량 자료를 가로로 합쳐서 연도별로 맥주, 소주에 부과된 세금과 이에 따른 판매량이 어떻게 변했는지 한눈에 볼 수 있게 했다.

### (5)분석 결과 도출:

```python
alcohol_df['결과'] = (alcohol_df['소주 부과세금'].diff() > 0 ) & (alcohol_df['소주 판매량'].diff() < 0) & (alcohol_df['맥주 판매량'].diff() > 0)

print('전체:', alcohol_df['결과'].count())
print('True:', alcohol_df['결과'].sum())
print('소주 가격변화로 확인한 경우 True의 비율',  alcohol_df['결과'].sum()/alcohol_df['결과'].count())

alcohol_df['결과'] = (alcohol_df['맥주 부과세금'].diff() > 0 ) & (alcohol_df['소주 판매량'].diff() > 0) & (alcohol_df['맥주 판매량'].diff() < 0)
print('전체:', alcohol_df['결과'].count())
print('True:', alcohol_df['결과'].sum())
print('맥주 가격변화로 확인한 경우 True의 비율',  alcohol_df['결과'].sum()/alcohol_df['결과'].count())
```

### (6)결론:
True의 비율이 현저히 낮다.<br>
소주와 맥주는 다른 한 재화의 가격이 올라도 판매량이 늘지 않는다. <br>
따라서 내가 세운 가설은 틀렸다. <br>
소주와 맥주는 대체재가 아니다.

### (10)소스코드

```python
import pandas as pd

alcohol_Pb=pd.read_csv("C://fightWithPython//PriceOfAlcohol.csv", encoding='cp949')
alcohol_P = alcohol_Pb[['연도', '맥주', '희석식 소주']].copy()
alcohol_P.rename(columns={alcohol_P.columns[1]:'맥주 부과세금'},inplace=True)
alcohol_P.rename(columns={alcohol_P.columns[2]:'소주 부과세금'},inplace=True)

alcohol_Qb= pd.read_csv("C://fightWithPython//QuantityOfAlcohol.csv", encoding='cp949')
alcohol_Q = alcohol_Qb[['맥주','희석식소주']].copy()
alcohol_Q.rename(columns={alcohol_Q.columns[0]:'맥주 판매량'},inplace=True)
alcohol_Q.rename(columns={alcohol_Q.columns[1]:'소주 판매량'},inplace=True)

alcohol_df=pd.concat([alcohol_P, alcohol_Q.reset_index(drop=True)], axis=1)

alcohol_df
```
